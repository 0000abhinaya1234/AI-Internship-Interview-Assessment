```python
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
```

```python
detector = dlib.get_frontal_face_detector()
```

```python
def apply_mask_to_face(image, mask_img, predictor, detector):
    img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    faces = detector(img_rgb)

    for face in faces:
        landmarks = predictor(img_rgb, face)

        # Get relevant points: nose, chin
        nose_point = (landmarks.part(30).x, landmarks.part(30).y)
        chin_bottom_point = (landmarks.part(8).x, landmarks.part(8).y)
        chin_left_point = (landmarks.part(4).x, landmarks.part(4).y)
        chin_right_point = (landmarks.part(12).x, landmarks.part(12).y)

        # Calculate width and height for the mask
        mask_width = int(np.linalg.norm(np.array(chin_right_point) - np.array(chin_left_point)) * 1.1)
        mask_height = int(np.linalg.norm(np.array(nose_point) - np.array(chin_bottom_point)) * 1.5)

        # Resize the mask
        resized_mask = cv2.resize(mask_img, (mask_width, mask_height))

        # Calculate center position
        center_x = nose_point[0]
        center_y = int((nose_point[1] + chin_bottom_point[1]) / 2)

        top_left_x = center_x - mask_width // 2
        top_left_y = center_y - mask_height // 2

        # Separate alpha channel
        if resized_mask.shape[2] == 4:
            mask_alpha = resized_mask[:, :, 3] / 255.0
            mask_rgb = resized_mask[:, :, :3]

            for c in range(0, 3):
                y1, y2 = max(0, top_left_y), min(image.shape[0], top_left_y + mask_height)
                x1, x2 = max(0, top_left_x), min(image.shape[1], top_left_x + mask_width)

                mask_y1 = max(0, -top_left_y)
                mask_y2 = mask_y1 + y2 - y1

                mask_x1 = max(0, -top_left_x)
                mask_x2 = mask_x1 + x2 - x1

                if y1 < y2 and x1 < x2:
                    image[y1:y2, x1:x2, c] = (
                        image[y1:y2, x1:x2, c] * (1 - mask_alpha[mask_y1:mask_y2, mask_x1:mask_x2]) +
                        mask_rgb[mask_y1:mask_y2, mask_x1:mask_x2, c] * mask_alpha[mask_y1:mask_y2, mask_x1:mask_x2]
                    )
    return image
```

```python
for root, dirs, files in os.walk(input_folder):
    for file_name in files:
        if file_name.lower().endswith(('.jpg', '.jpeg', '.png')):
            img_path = os.path.join(root, file_name)
            img = cv2.imread(img_path)

            if img is None:
                print(f"❌ Error reading {file_name}, skipping...")
                continue

            masked_img = apply_mask_to_face(img, mask_img, predictor, detector)

            if masked_img is not None:
                # Save in masked_faces keeping same folder structure
                relative_path = os.path.relpath(img_path, input_folder)
                output_path = os.path.join(output_folder, relative_path)

                os.makedirs(os.path.dirname(output_path), exist_ok=True)
                cv2.imwrite(output_path, masked_img)
                print(f"✅ Masked image saved: {output_path}")
            else:
                print(f"❌ Face not detected in {file_name}, skipping...")
```
